```
import numpy as np
import os
import matplotlib as mpl
import matplotlib.pyplot as plt

import rasterio as rio
import geopandas as gpd
import earthpy as et
import earthpy.spatial as es
```

```
mpl.rcParams['figure.figsize'] = (10, 10)
mpl.rcParams['axes.titlesize'] = 20
```

```
with rio.open("m_3910505_nw_13_1_20150919_crop.tif") as src:
    naip_data = src.read()
    
# View shape of the data
naip_data.shape
```

```
naip_data_meta=src.meta
```

```
fig, ax = plt.subplots()
ax.imshow(naip_data[0], cmap='Greys')
ax.set(title="NAIP RGB Imagery - Band 1-Red\nCold Springs Fire Scar",
       xticks=[], yticks=[])
plt.show()

```

```
fig, ax = plt.subplots()
ax.imshow(naip_data[2], cmap='Greys')
ax.set(title="NAIP RGB Imagery - Band 3-blue\nCold Springs Fire Scar",
       xticks=[], yticks=[])
plt.show()
```

```
fig, ax = plt.subplots()
ax.imshow(naip_data[3], cmap='Greys')
ax.set(title="NAIP RGB Imagery - Band 4-NIR\nCold Springs Fire Scar",
       xticks=[], yticks=[])
plt.show()
```

```
FDI = (naip_data[3] -(naip_data[0] + naip_data[2]))
```

```
# Plot FDI data
fig, ax = plt.subplots(figsize=(12,6))
fdii = ax.imshow(FDI, cmap='PiYG',
                vmin=-1, vmax=1)
fig.colorbar(fdii, fraction=.05)
ax.set(title="NAIP Derived FDI\n 19 September 2015 - Cold Springs Fire, Colorado")
ax.set_axis_off()
plt.show()
```
